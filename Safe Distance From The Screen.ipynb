```python
import cv2
from cvzone.FaceMeshModule import FaceMeshDetector

cap = cv2.VideoCapture(0)
detector = FaceMeshDetector(maxFaces=1)
# Set the window size
window_width = 1280
window_height = 720
cv2.namedWindow("Image", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Image", window_width, window_height)

while True:
    _, frames = cap.read()

    frames, faces = detector.findFaceMesh(frames, draw=False)

    text = "Safe distance maintained."

    if faces:
        face = faces[0]
        right_eye = face[145]
        left_eye = face[374]

        # Calculate the width of the eyes(constant value)
        W = 6.3
        w, _ = detector.findDistance(right_eye, left_eye)

      
        f = 837  # after being calculated 
        d = (W * f) / w

        print(f"Distance: {d:.2f} units")  

        if d < 75:
            text = "Please keep a safe distance."

    # Display the text on the frame
    cv2.putText(frames, text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (240,55, 99), 2)

    cv2.imshow("Image", frames)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
